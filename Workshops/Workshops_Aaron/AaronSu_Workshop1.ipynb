## Exercise 4: Dots and Lines

Time to get some plotting practice. Create three different plots of 100 values between 0-20, one displaying a linear function, one displaying a quadratic function, and one displaying a quartic function. Make sure each plot is properly labeled and use a different method to display the data for each of the three plots. 

```python
import numpy as np
import matplotlib.pyplot as plt


#Linear Equation
x = np.linspace(0,3,100)
y = 3*x
plt.plot(x,y,'r.')
plt.title('Linear Equation')
plt.xlabel('X Value' )
plt.ylabel('Y Value' )
plt.ylim(0,100)
plt.xlim(0,np.pi)
plt.show()

#Quadratic Equation
x6 = np.linspace(0,3,100)
y6 = 6*x**2
plt.plot(x6,y6,"m-")
plt.title('Quadratic Equation')
plt.xlabel('X Value' )
plt.ylabel('Y Value' )
plt.ylim(0,100)
plt.xlim(0,np.pi)

plt.show()

#Quintic Equation
x7 = np.linspace(0,3,100)
y7 = x**4+3*x*2+np.sqrt(13)*x+5
plt.plot(x7,y7,"g.-")
plt.title('Qunitic Equation')
plt.xlabel('X Value' )
plt.ylabel('Y Value' )
plt.ylim(0,100)
plt.xlim(0,np.pi)


plt.show()
```

## Exercise 5: Sine-ster Plots

In this next exercise you are going to create four different plots of sine wave functions. Your four plots should include; one fairly standard sine wave beginning at x = 0 and y = 0, one which has an adjusted x-axis to show both positive and negative aspects of the wave, one in which you adjust the amplitude of the wave but have changed the y-axis ticks so that it matches your other plots, and finally display a cosine and a sine wave on the same plot distinguishing them as distinctly as you can and making sure to properly label the figure so the reader knows which is which. 

```python
from pylab import *

# make plots inline
%matplotlib inline


#Standard Sine Wave
x0 = arange(0,2*pi,pi/100)
plot(x0,sin(x0),'-',)
xlabel('X Value' )
ylabel('Y Value' )
plt.title ('Standard Sine Wave')
show()

#Negative and Positive Sine Wave
x1 = arange(-pi,pi,pi/100)
plot(x1,sin(x1),'-',)
xlabel('X Value' )
ylabel('Y Value' )
plt.title('Negative and Postitive Sine Wave')
show()

#Changed Amplitude Sine Wave
x2 = arange(-pi,pi,pi/100)
plot(1.25*x2,sin(x2),'-',)
xlabel('X Value' )
ylabel('Y Value' )
ylim(-1,1)
xlim(-3,3) #Used limits to enforce the steps
plt.title('Amplitude Sine Wave')
show()

#Sine Wave and Cosine Wave
x4 = arange(-pi, pi, pi/100)   # create an array of x values
plot(1.5* x4, sin(x4), label='Sine' )   # plot arrays x and sin(x), in blue, solid line
plot(x4, cos(x4),label='Cosine' )
xlabel('X Value' )
ylabel('Y Value' )
xlim(-pi, pi)
ylim(-1, 1)
legend(loc='lower right')
plt.title('Sine Wave and Cosine Wave')
show()
```

## Exercise 6: Kinematics Plot

Create a single figure that shows separate graphs of vertical position, velocity, and acceleration for an object in free-fall. Your plot should have a single horizontal time axis and separate stacked graphs showing position, velocity, and acceleration each on their own vertical axis. The online [matplotlib gallery](http://matplotlib.org/gallery.html) will probably be helpful. Make each curve a different color.

Hint: Look for "subplots_demo" or "shared_axis_demo" in the gallery.

```python
import numpy as np
import matplotlib.pyplot as plt

h = 1000 #Height, in meters
g = 9.8 #Gravity, in meters squared
t = np.linspace(0,np.sqrt(2*h/g),100) 
y = h - 0.5*g* (t**2) 
v = g*t
a = [9.8]*100


f, axarr = plt.subplots(3, sharex=True)
axarr[0].plot(t, y,'m')
axarr[0].set_ylabel('Altitude (m)')
axarr[1].plot(t, v,'b')
axarr[1].set_ylabel('Velocity (m/s)')
axarr[2].plot(t, a,'g')
axarr[2].set_ylabel('Acceleration(m/s)')
axarr[0].set_title('Figure In Free Fall')
axarr[2].set_xlabel('Time (s)')
plt.show()
```
